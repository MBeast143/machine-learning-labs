## DATA PREPARATION

```python
data_prepared_tbl <- bitcoin_prices_tbl %>%
  select(Datetime, `Adj Close`)

splits <- time_series_split(
  data_prepared_tbl,
  assess = 5,
  cumulative = TRUE
)

splits
```

## MODELING

```python
wflw_fift <- workflow() %>%
  add_model(
  arima_reg() %>% set_engine("auto_arima")
  ) %>%
    add_recipe(
    recipe = recipe(`Adj Close` ~ Datetime, training(splits))
    ) %>%
      fift(training(splits))
```

```python
model_tbl <- modeltime_table(
  wflw_fift
)
```
